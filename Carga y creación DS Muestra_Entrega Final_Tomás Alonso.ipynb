
# Carga API

```python
import requests
import json
import pandas as pd
url='https://api.worldbank.org/pip/v1/pip?country=all&year=2018&povline=1.9&fill_gaps=false&format=json'
df= requests.get(url)
print(df)
```

```python
print(len(df.text))
df.text
```

```python
texto=df.text
json.loads(texto)
```

```python
jsondata=json.loads(texto)
df2=pd.DataFrame.from_dict(jsondata)
df2.head()
```

## Segunda API 

```python
url4='https://api.worldbank.org/pip/v1/pip-grp?country=all&year=2018&povline=1.9&group_by=wb&format=json'
agdata= requests.get(url4)
print(agdata)
```

```python
print(len(agdata.text))
agdata.text
```

```python
texto2=agdata.text
json.loads(texto2)
```

```python
jsondata2=json.loads(texto2)
df3=pd.DataFrame.from_dict(jsondata2)
df3.head(10)
```

## Unión de API's

```python
dfm=pd.merge(df2,df3,on="region_name")
dfm
```

```python
dfm.info()
```

```python
dfm.isnull().sum()
```

```python
dfm.dtypes
```

### Visualización de nulos

```python
import missingno as msno
msno.matrix(dfm)
```

### Reemplazo de booleanos

```python
bool_col = dfm.select_dtypes(include=['bool'])
for col in bool_col.columns.values:
    dfm[col] = dfm[col].astype('int64')
```

# Creación DataSet Pobreza Regional

```python
pob=pd.read_csv('train_values_wJZrCmI.csv')
prob_pob=pd.read_csv('train_labels.csv')
```

```python
pob.head()
```

### Relleno de nulos 

```python
pob['bank_interest_rate'].fillna(0,inplace=True)
pob['mm_interest_rate'].fillna(0,inplace=True)
pob['mfi_interest_rate'].fillna(0,inplace=True)
pob['other_fsp_interest_rate'].fillna(0,inplace=True)
pob['education_level'].fillna(pob['education_level'].mean(), inplace=True)
pob['share_hh_income_provided'].fillna(pob['share_hh_income_provided'].mean(),inplace=True)
```

```python
print(pob.isnull().sum())
```

### Reemplazo de booleanos y flotantes

```python
bool_col = pob.select_dtypes(include=['bool'])
for col in bool_col.columns.values:
    pob[col] = pob[col].astype('int64')
```

```python
float_col = pob.select_dtypes(include=['float64'])
for col in float_col.columns.values:
    pob[col] = pob[col].astype('int64')
```

```python
pob.info()
```

## Unión dataset probabilidad

```python
pob1=pob.merge(prob_pob, on='row_id')
```

```python
pob1=pob1.set_index('row_id')
```

### Cambio label datos

```python
pob1['country'].replace({'J':'Sub-Saharan Africa','D':'Europe & Central Asia', 'I':'Other High Income Countries',
                        'G':'Latin America & Caribbean', 'A':'South Asia', 'C':'Middle East & North Africa', 'F':'East Asia & Pacific'}, inplace=True)
```

```python
pob1.head()
```

### Label Encoder DS Probabilidad

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
pob1['religion']= label_encoder.fit_transform(pob1['religion'])
  
pob1['religion'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
pob1['relationship_to_hh_head']= label_encoder.fit_transform(pob1['relationship_to_hh_head'])
  
pob1['relationship_to_hh_head'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
pob1['employment_category_last_year']= label_encoder.fit_transform(pob1['employment_category_last_year'])
  
pob1['employment_category_last_year'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
pob1['employment_type_last_year']= label_encoder.fit_transform(pob1['employment_type_last_year'])
  
pob1['employment_type_last_year'].unique()
```

#### Cambio nombre variable region DS API

```python
dfm.rename(columns = {'region_name':'country'}, inplace = True)
```

```python
dfm.head()
```

# Unión DS API con DS Probabilidad

```python
pob2=pob1.merge(dfm, on='country')
```
